```python
import os
import cv2
from glob import glob as gg
from os.path import join as pjoin
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
bpath = pjoin('/Users', 'noemieter', 'Documents', 'LaCNS', 'repositories', 'LaCNS_website')
path = pjoin(bpath, 'assets', 'images', 'headshots_blackwhite')
os.listdir(path)
```

```python
for file in os.listdir(path):
    if not file.startswith('.'):
        fig, axes = plt.subplots()
    
        # read the image file
        image = cv2.imread(pjoin(path, file))
        
        # add gray effect
        gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        
        # invert the image
        inverted_image = cv2.bitwise_not(gray_image)
        
        # blur the image
        blur_image = cv2.GaussianBlur(inverted_image, (111, 111), 0)
        
        # create inverted blur image
        inverted_blur = cv2.bitwise_not(blur_image)
        
        # divide the gray image and inverted blur
        sketch = cv2.divide(gray_image, inverted_blur, scale=256.0)
        
        # save the file to the folder
        cv2.imwrite(pjoin(path, 'new_' + file), sketch)
        
        axes.imshow(sketch, cmap='gray')
        axes.set_axis_off()
```

```python
_, axes = plt.subplots()
# read the image file
image = cv2.imread(pjoin(path, 'pic_FB.jpg'))

# add gray effect
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# invert the image
inverted_image = cv2.bitwise_not(gray_image)

# blur the image
blur_image = cv2.GaussianBlur(inverted_image, (111, 111), -2)

# create inverted blur image
inverted_blur = cv2.bitwise_not(blur_image)

# divide the gray image and inverted blur
sketch = cv2.divide(gray_image, inverted_blur, scale=200.0)

# save the file to the folder
cv2.imwrite("sketch.png", sketch)

axes.imshow(sketch, cmap='gray')
axes.set_axis_off()
```
